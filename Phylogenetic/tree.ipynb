```python
from Bio import Phylo
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import pandas as pd
import re
```

```python
meta_data_df = pd.read_excel("420 FHB_PGRC-BreadW_Master File.xlsx", sheet_name="Sheet2")
def loc_finder(id):
    loc = meta_data_df[meta_data_df["Cross"] == id]["Origin"]
    if loc.empty:
        return None
    else:
        return re.sub(r'^\d*    ', '', loc.to_string())
    
def dis_finder(id):
    severity = meta_data_df[meta_data_df["Cross"] == id]["Severity"]
    if severity.empty or severity.item() <= 10.0:
        return "black"
        
    else:
        return "red"
```

```python
Phylo.convert("hapmaps/wheat_hapmap_gen_upgma.nwk", "newick", "UPGMA_old.xml", "phyloxml")
tree = Phylo.read('UPGMA_old.xml', 'phyloxml')
tree = tree.as_phyloxml()
label_col = {}
i = 0
k = 0
for child in tree.get_terminals():
    k = k + 1
    if str(child.name) != "None" and str(child.name) != "�" :
        child.name = child.name[0:2] + " " + child.name[2:] # To align with cross formatting
        i = i + 1
        old_name = child.name 
        child.color = dis_finder(child.name)
        child.name = child.name + " (" + str(loc_finder(child.name)) + ")"
        label_col[child.name] = dis_finder(old_name)
            
print(f"Encorperated {i} taxa of {k}")           
plt.rcParams["figure.figsize"] = (150,150)

# For unimplemented legend
infect_lbl = mpatches.Patch(color='red', label='Infected Individual')
health_lbl = mpatches.Patch(color='black', label='Healthy Individual')

Phylo.draw(
    tree,
    label_colors = label_col
)
```
