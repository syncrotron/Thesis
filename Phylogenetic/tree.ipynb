```python
from Bio import Phylo
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import pandas as pd
import re
```

```python
meta_data_df = pd.read_excel("420 FHB_PGRC-BreadW_year2020.xlsx", sheet_name="Sheet3")
def loc_finder(id):
    loc = meta_data_df[meta_data_df["new ID"] == id]["Origin"]
    if loc.empty:
        return None
    else:
        return re.sub(r'^\d*    ', '', loc.to_string())
    
def dis_finder(id):
    if meta_data_df[meta_data_df["new ID"] == id]["20-R1 FDK%"].item() >= 10.0:
        return "red"
    else:
        return "black"
```

```python
Phylo.convert("UPGMA_old.dnd", "newick", "UPGMA_old.xml", "phyloxml")
tree = Phylo.read('UPGMA_old copy.xml', 'phyloxml')
tree = tree.as_phyloxml()
label_col = {}

for child in tree.get_terminals():
    if str(child.name) != "None" and str(child.name) != "ï¿½":
            old_name = child.name 
            child.color = dis_finder(child.name)
            child.name = child.name + " (" + loc_finder(child.name) + ")"
            label_col[child.name] = dis_finder(old_name)
            
            
plt.rcParams["figure.figsize"] = (150,150)

# For unimplemented legend
infect_lbl = mpatches.Patch(color='red', label='Infected Individual')
health_lbl = mpatches.Patch(color='black', label='Healthy Individual')

Phylo.draw(
    tree,
    label_colors = label_col
)
```
